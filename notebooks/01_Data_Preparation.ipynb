# Data Preparation

This notebook focuses on preparing the Hyperliquid trader dataset and the Bitcoin Fear & Greed sentiment dataset for analysis. The preprocessing steps include data inspection, handling missing values and duplicates, converting data types, standardizing date formats, and merging both datasets into a single analysis-ready dataset.

## Objectives

The objectives of this notebook are to:

- Load the trader and market sentiment datasets.
- Inspect the structure and quality of each dataset.
- Check for missing values and duplicate records.
- Convert timestamp columns to appropriate datetime formats.
- Align both datasets using a common daily date.
- Merge trader data with the corresponding daily market sentiment.
- Validate the merged dataset before exploratory data analysis.

### Data Type Conversion

Several columns required conversion to appropriate data types before analysis.

- The `Timestamp IST` column was converted from string to datetime.
- The sentiment dataset's `date` column was converted to datetime.
- Numeric columns such as execution price, trade size, fee, and profit/loss were verified to have appropriate numeric data types.

Using correct data types enables accurate calculations, filtering, and time-series analysis.

```python
trader_df.dtypes
```

```python
trader_df["Timestamp IST"] = pd.to_datetime(
    trader_df["Timestamp IST"],
    format="%d-%m-%Y %H:%M"
)
```

```python
trader_df["Timestamp IST"].dtype
```

### Date Standardization

The trader dataset records transactions at the timestamp level, whereas the market sentiment dataset provides one sentiment value per day.

To align both datasets, only the date component was extracted from the trader timestamps. This created a common daily date field that could be used to merge both datasets.

```python
trader_df["date"] = trader_df["Timestamp IST"].dt.date
```

```python
sentiment_df["date"] = pd.to_datetime(
    sentiment_df["date"],
    format="%d-%m-%Y"
)
```

```python
sentiment_df["date"] = sentiment_df["date"].dt.date
```

### Merging the Datasets

The trader dataset was merged with the market sentiment dataset using the common daily date.

A left join was performed to preserve all trading records while attaching the corresponding daily market sentiment (Fear or Greed) to each trade.

This merged dataset forms the foundation for subsequent exploratory data analysis and trader performance evaluation.

```python
merged_df = pd.merge(
    trader_df,
    sentiment_df,
    on="date",
    how="left"
)
```

### Merge Validation

After merging, the resulting dataset was validated by:

- Verifying the number of records
- Checking for unmatched sentiment values
- Inspecting data types
- Confirming successful alignment of daily market sentiment with each trading record

The final dataset is now ready for feature engineering and exploratory data analysis.
